```python
#% del total que suponen los valores faltantes
def porcentaje_missings(data):
    m_per = data.isnull().sum() * 100 / len(data)
    percent_missing = round(m_per,2)
    missing_value_df = pd.DataFrame({'percent_missing (%)': percent_missing})
    return missing_value_df

porcentaje_missings(df)
```

```python
#% del total que suponen los valores faltantes en las filas
def porcentaje_missings_filas(data):
    m_per = data.isnull().sum(axis=1) * 100 / len(data)
    percent_missing = round(m_per,2)
    missing_value_df = pd.DataFrame({'percent_missing (%)': percent_missing})
    return missing_value_df
porcentaje_missings_filas(df)
```

```python
import matplotlib.pyplot as plt # Importamos la librería matplotlib. 
# El inline es para que se muestren los gráficos en un jupyter notebook
%matplotlib inline 
```

```python
import seaborn as sns # Importamos la librería Seaborn
sns.set(style='darkgrid') # Especificamos un fondo gris de rejilla
```

```python
bins = [0, 20, 40, 60, 80, 100, 120]
names = ["0-19", "20-39", "40-59", "60-79", "80-99", "100-119"]
df_copy["Edad"] = pd.cut(df_copy["Edad"], bins, labels = names)
```

```python
df_copy.head()
```

```python
df_copy.groupby('Edad').mean()
```

```python
# Diagramas de barras de dos factores de frecuencias condicionadas a los grupos de la variable x

x = 'Edad'
y = 'Sexo'


pct2 = (df_copy.groupby([x,y]).size() / df_copy.groupby([x]).size()).reset_index().rename({0:'Frecuencia Condicionada'}, axis=1)


# seaborn
sns.barplot(x=x, hue=y, y='Frecuencia Condicionada', data=pct2)
plt.show()

pct2
```
